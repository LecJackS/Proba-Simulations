```r
# b) Repetir para una Gamma(5, 1/2) 
# rgamma(n, shape, rate = 1, scale = 1/rate)
muestra_25 <- rgamma(25, shape=5, rate = 1/2)
muestra_50 <- rgamma(50, shape=5, rate = 1/2)
muestra_100 <- rgamma(100, shape=5, rate = 1/2)
muestra_10000 <- rgamma(10000, shape=5, rate = 1/2)
```

```r
# QQ-Plots
par(mfrow=c(2,2))
qqnorm(muestra_25, main="n=25")
qqnorm(muestra_50, main="n=50")
qqnorm(muestra_100, main="n=100")
qqnorm(muestra_10000, main="n=10000")
```

```r
# c) Repetir para una Y = Z / U 
# Z ~ N(0,1)
muestra_Z_25 <- rnorm(25)
muestra_Z_50 <- rnorm(50)
muestra_Z_100 <- rnorm(100)
muestra_Z_10000 <- rnorm(10000)
# U ~ U(0,1)
muestra_U_25 <- runif(25)
muestra_U_50 <- runif(50)
muestra_U_100 <- runif(100)
muestra_U_10000 <- runif(10000)
# Y = Z / U 
muestra_Y_25 <- muestra_Z_25 / muestra_U_25
muestra_Y_50 <- muestra_Z_50 / muestra_U_50
muestra_Y_100 <- muestra_Z_100 / muestra_U_100
muestra_Y_10000 <- muestra_Z_10000 / muestra_U_10000
```

```r
# QQ-Plots: Y ~ Norm/Unif
par(mfrow=c(2,2))
qqnorm(muestra_Y_25, main="n=25")
qqnorm(muestra_Y_50, main="n=50")
qqnorm(muestra_Y_100, main="n=100")
qqnorm(muestra_Y_10000, main="n=10000")
```

```r
# d)
# QQ-Plots: U ~ Uniforme
par(mfrow=c(2,2))
qqnorm(muestra_U_25, main="n=25")
qqnorm(muestra_U_50, main="n=50")
qqnorm(muestra_U_100, main="n=100")
qqnorm(muestra_U_10000, main="n=10000")
```

Media $α$-Podada:  es  un  promedio  calculado  sobre  los  datos  una  vez  que  se  han eliminado $α$ datos de cada extremo.

Es una medida intermedia entre la media y la mediana.

https://es.wikipedia.org/wiki/Media_truncada

```r
# Ejemplo extraigo uno de cada lado:
A <- c(3,4,6,2,8,4,3,4,5,9)
# Ordeno
A <- c(sort(A))
# Alfa = 10%
alfa <- length(A)/10
A
A[(alfa+1):(length(A)-alfa)]
```

```r
n <- length(X[,])
alfa <- n/10
X_alfa_podado <- c(sort(X[,]))[(alfa+1):(n-alfa)]
media_alfa_podada <- mean(X_alfa_podado)
```
